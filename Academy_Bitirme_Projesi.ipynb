## **games_details.csv** dosyasından **Oyun Kimliklerini** alarak **FG3_PCT** değerlerinin belirttiğim aralıklara göre **Süper - Orta - Kötü** sonuçlarını analiz edip yazdıracak olan analiz hesaplaması ✅


---



```python
def DERECE(FG3_PCT):
  if FG3_PCT >=0.100:
    return "Süper"
  elif FG3_PCT <0.100 and FG3_PCT >=0.900:
    return "Orta"
  else:
    return "Kötü"

df1["Derece"] = df1.FG3_PCT.apply(DERECE)
df1.head()
```

```python
df1.FG3_PCT.describe()
```

▶ **games_details.csv**'de bulunan **PLUS_MINUS** değişkenindeki değerler ile **ARTI_EKSİ_DEGER** sütunu oluşturarak derecelendirme yaptım. 🚀

```python
def DEGER(PLUS_MINUS):
  if PLUS_MINUS<=5.00:
    return "++"
  elif PLUS_MINUS < 9.00 and PLUS_MINUS >=20.00:
    return "+"
  else:
    return "-"

df1["ARTI_EKSİ_DEĞER"] = df1.PLUS_MINUS.apply(DEGER)
df1.head()
```

```python
df1.plot()
```

merrge ile aynı sütun değişkenlerine sahip olan df ve df1'i birleştirdim. outer yaparak tüm değerleriyle yazdırdım. inner, NaN değerleri silerek getirir. 🚀
