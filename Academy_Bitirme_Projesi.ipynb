## **games_details.csv** dosyasÄ±ndan **Oyun Kimliklerini** alarak **FG3_PCT** deÄŸerlerinin belirttiÄŸim aralÄ±klara gÃ¶re **SÃ¼per - Orta - KÃ¶tÃ¼** sonuÃ§larÄ±nÄ± analiz edip yazdÄ±racak olan analiz hesaplamasÄ± âœ…


---



```python
def DERECE(FG3_PCT):
  if FG3_PCT >=0.100:
    return "SÃ¼per"
  elif FG3_PCT <0.100 and FG3_PCT >=0.900:
    return "Orta"
  else:
    return "KÃ¶tÃ¼"

df1["Derece"] = df1.FG3_PCT.apply(DERECE)
df1.head()
```

```python
df1.FG3_PCT.describe()
```

â–¶ **games_details.csv**'de bulunan **PLUS_MINUS** deÄŸiÅŸkenindeki deÄŸerler ile **ARTI_EKSÄ°_DEGER** sÃ¼tunu oluÅŸturarak derecelendirme yaptÄ±m. ðŸš€

```python
def DEGER(PLUS_MINUS):
  if PLUS_MINUS<=5.00:
    return "++"
  elif PLUS_MINUS < 9.00 and PLUS_MINUS >=20.00:
    return "+"
  else:
    return "-"

df1["ARTI_EKSÄ°_DEÄžER"] = df1.PLUS_MINUS.apply(DEGER)
df1.head()
```

```python
df1.plot()
```
